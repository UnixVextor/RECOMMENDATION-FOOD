## Cosine Similarity 
**Cosine similarity** ‡∏´‡∏£‡∏∑‡∏≠ ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Ñ‡∏•‡∏∂‡∏á‡πÇ‡∏Ñ‡πÑ‡∏ã‡∏ô‡πå ‡∏Ñ‡πà‡∏≤‡∏à‡∏∞‡∏Ç‡∏∂‡πâ‡∏ô‡∏≠‡∏¢‡∏π‡πà‡∏Å‡∏±‡∏ö‡∏°‡∏∏‡∏°‡∏ó‡∏µ‡πà‡πÄ‡∏ß‡∏Å‡πÄ‡∏ï‡∏≠‡∏£‡πå 2 ‡πÄ‡∏ß‡∏Å‡πÄ‡∏ï‡∏≠‡∏£‡πå‡∏Å‡∏£‡∏∞‡∏ó‡∏≥‡∏ï‡πà‡∏≠‡∏Å‡∏±‡∏ô
![Cosine Similarity](image/cosine_similarity.png)

‡πÇ‡∏î‡∏¢‡∏ó‡∏µ‡πà Cosine Similarity ‡∏à‡∏∞‡∏°‡∏µ‡∏™‡∏π‡∏ï‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ
### $$D_{Cosine} = cos\theta = \dfrac{u\cdot v}{|u||v|}$$

- $u$ : ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 1
- $v$ : ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 2


‡∏à‡∏≤‡∏Å‡∏™‡∏°‡∏Å‡∏≤‡∏£‡∏ô‡∏≥‡∏°‡∏≤‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÄ‡∏õ‡πá‡∏ô‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡∏ú‡∏•‡∏£‡∏ß‡∏°‡πÑ‡∏î‡πâ‡∏Ñ‡∏∑‡∏≠
### $$D_{Cosine} = \dfrac{\sum_{i=1}^{n} A_i B_i}{\sqrt{\sum_{i=1}^{n} A_i^2} \sqrt{\sum_{i=1}^{n} B_i^2}}$$



```python
import math
def cosinSimilarity(lstInput):
    for i in range(233):
        data = np.array([Filter_df['Energy(kcal)'][i],Filter_df['Protein(g)'][i],Filter_df['Fat(g)'][i],Filter_df['Carbohydrate, available\n(carbohydrate, total)(g)'][i]],dtype=float)
        dot = np.dot(lstInput,data)
        magData = np.linalg.norm(data)
        magLstInput = np.linalg.norm(lstInput)
        cosine = dot/(magData*magLstInput)
        degree = float("{0:.3f}".format((math.acos(cosine) * 180/math.pi)))
        
#         print("Cosine =","{0:.6f}".format(cosine),end=' ') 
#         print("Degree =",degree,'¬∞')
        Filter_df.loc[i,'Cosine Similarity'] = "{0:.6f}".format(cosine)
        Filter_df.loc[i,'Degree Cosine Similarity'] = degree
```

### ‚å®Ô∏è ‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ

```python
energy = float(input("Enter Energy(kcal) : "))
protein = float(input("Enter Protein(g) : "))
fat  = float(input("Enter Fat(g) : "))
carbohydrate = float(input("Enter Carbohydrate(g) : "))
lstInput = np.array([energy,protein,fat,carbohydrate])
# printDetail(energy,protein,fat,carbohydrate)
cosinSimilarity(lstInput)
Filter_df
```

### Correlation Matrix
  Correlation ‡∏´‡∏£‡∏∑‡∏≠ ‡∏™‡∏´‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡∏Ñ‡∏∑‡∏≠‡∏Å‡∏≤‡∏£‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏´‡∏≤‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ 2 ‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏Ç‡∏∂‡πâ‡∏ô‡πÑ‡∏õ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡∏™‡∏°‡∏Å‡∏≤‡∏£‡∏Ñ‡∏∑‡∏≠ 
   ### $$r_{xy} = \dfrac{\sum (x_i-\bar x)(y_i-\bar y)}{\sqrt{\sum (x_i-\bar x)^2 \sum (y_i-\bar y)^2}}$$
   
- $x$ : ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 1
- $y$ : ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 2
- $\bar x $ : ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 1
  $$ \bar x = \dfrac{\sum_{i=1}^{n} x_i}{n}$$
- $\bar y $ : ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 2
  $$ \bar y = \dfrac{\sum_{i=1}^{n} y_i}{n}$$

```python
cor = Sort_df.iloc[:10,1:].transpose().corr()
cor
```
